```python
import bq_time_check_run
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
bytecode_sql = 'SELECT * FROM `capture_milla.bytecode_descriptives_rank`'
opcode_sql = 'SELECT * FROM `capture_milla.opcode_descriptives_rank`'

bytecode_rank, _ = run_query(bytecode_sql)
opcode_rank, _ = run_query(opcode_sql)
```

```python
fig1, bytep = plt.subplots(figsize=(16,7))

bytep = bytecode_rank.plot(kind='scatter', x='rank', y='n', ax=bytep)
bytep.set_xscale('log')
bytep.set_yscale('log')
bytep.title.set_text('Bytecode Cluster rank vs Cluster size')
bytep.set_ylabel('Cluster size')
bytep.set_xlabel('Cluster rank')
bytep.get_figure().savefig('G5 bytecode cluster rank vs cluster size.pdf')
```
